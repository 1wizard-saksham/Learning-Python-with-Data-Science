# Indexing and Slicing of NumPy array

```python
# Creating an array of prime numbers

x4 = np.array([2,3,5,7,11,13,17,19])
print(x4[3])         # Indexing in NumPy array
print(x4[0:5])       # Slicing the NumPy array
print(x4.dtype)
print(x4.shape)
```

# Difference between list and NumPy array

- A NumPy array is faster than Python lists
- NumPy array supports homogenous datatypes but lists support multiple datype
- NumPy array are memory efficient than Python lists

<span style="color: red;">PROOF:-</span>

# NumPy array supports homogenous datatypes but lists support multiple datype

```python
li = [2,4,'Saksham', True]          # Multiple datatypes. Hence, Heterogenous
# Creating an array from the list
li_array = np.array(li)
li_array                # Treats each element of the as a string because these elements can be represented as a string. Hence, Homogenous
```

# NumPy array are memory efficient than Python lists


```python
ones = np.array([1,1,1,1,1,1])
ones.dtype
```

```python
# Controlling the memory consumption in NumPy array unlike lists

ones = np.array([1,1,1,1,1,1], dtype = np.int8)
ones.dtype
```

# A NumPy array is faster than Python lists

```python
import random as r
r.seed(2003)
test_list = [r.randint(1,10) for i in range(50000)]
print(f"list= {test_list}")
# Creating an array frm the list
test_list_array = np.array([test_list])
print(f"array= {test_list_array}")
```

```python
# Checking the execution time 


sq = [item**2 for item in test_list]
sq
```

```python
# Checking the execution time


sq_array = [item1**2 for item1 in test_list_array]
sq_array
```

    Hence, the exeution time of an array is greater thean that of list

<span style="color: red;">PROVED</span>

  Besides creating an array from a sequence of elements, you can create an array filled with 0's, 1's, empty depending on the state of the memory

```python
# 1) Creating an array containing 5 ones

one_array= np.ones(5)
print(one_array)
print(one_array.dtype)
print(len(one_array))           # Hence, the length of the array is 5
```

```python
# 2) Creating an array containing zeros

zero_array = np.zeros(5, dtype = np.float32)
print(zero_array)
print(zero_array.dtype)
print(len(zero_array)) 
```

```python
# 3) Creating an empty array
empty_array = np.empty(10)
print(empty_array)              # However, prints out garbage value depending upon the state of the memory
```
