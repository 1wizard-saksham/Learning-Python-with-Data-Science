```python
import numpy as np
```

## i) Creating a NumPy array through "ARANGE"  (Same as range function)

```python
numbers = np.arange(1,10,1)      # [Start, End, Step]
numbers
```

# ii) Linspace (linear spacing)

```python
x = np.linspace(0,1,5)      #  [0 dekhi 1 samma 5 numbers with a common difference ]
y = np.linspace(10,100,2)   #  [10 dekhi 100 samma 2 ota numbers with a common difference]
print(x)
print(y)
```

# iii) Sin in NumPy

```python
x1 = np.linspace(0,10,50)
sinx = np.sin(x1)
cosx = np.cos(x1)
import matplotlib.pyplot as plt
plt.plot(x1, sinx,  'o--', color = 'purple', label = 'sinx')  # o- = point and line, o-- = point and dotted line 
plt.plot(x1, cosx,  'o--', color = 'blue', label='cosx')
plt.legend()     # Yo lekhena vane label dekhaudaina
plt.show()     # Yo lekhena vane <matplotlib.legend.Legend at 0x1c9f2b89d30> aauxa
```

# iv) np.any() vs np.all()

```python
x2 = np.array([1,2,3,4,5,6,6,5,4,3,2,1])
```

```python
x2>=5  # No need to iterate unlike lists. So, efficient than list
# Returns Boolean
```

```python
x2[x2>=5]           # Slicing elements in x where x>=5. True vako indices ko value retrive garxa
```

```python
np.any(x2>=5)      # Returns boolean.  "If 'any' element >=5 exists". Like OR
```

```python
np.all(x2>=5)      # Like AND. Returns True if all the values are >=5
```

# v) Bitwise Operator (NumPy doesn't support logical operators)

### a) Bitwise AND (&)

```python
x2 = np.array([1,2,3,4,5,6,6,5,4,3,2,1])
# x[x>=3 and x<5] . This line is invalid
x2[(x2>=3) & (x2<5)]
```

### b) Bitwise OR (|)

```python
x2[(x2>=3) | (x2<5)]
```

# vi) Vector operation in array (One dimensional array works as a vector)

```python
x4 = np.array([2,3])
y4 = np.array([1,9])
print(x4+y4)
print(x4-y4)
print((x4/y4))
print(x4*y4)
```

# vii) Generating a random array and data distribution

```python
# Generating integer
import numpy as np
random_number = np.random.randint(low=1, high=50)
# Adding seed
np.random.seed(4)  # Generates a same value everytime the program is run 
random_number
```

```python
# Generating random numbers between 1 to 10
np.random.seed(0)
random_number1 = np.random.randint(low=1, high = 10, size = 20)   # Hence, size of array can be determined
random_number1
```

## a) Generating Uniform Distribution

```python
uniform_dist = np.random.uniform(low=2, high = 5, size = 1500)         # 2 dekhi 5 samma 1500 ota data
np.random.seed(0)
uniform_dist
```

## b) Generating Normal Distribution

```python
normal_dist = np.random.normal(loc = 5, scale = 1, size = 5000)     # loc = mean AND scale = Standard Deviation
np.random.seed(3)
normal_dist
```

# viii) Two Dimensional Array

```python
marks = [[2,3,4],[7,8,9],[11,12,13]]
marks_array = np.array(marks)
marks_array
marks_array[1][0]    #Indexing of 2D array. Prints out '7'
marks_array[1,0]     # Another way of Indexing
```

a) **Indexing first element of each row**

```python
marks_array[:,0]
```

 b) **Indexing out 11,12,13**

```python
marks_array[:2, 1:]
```

# ix) ReShape() method (Converting 1D to 2D array)

```python
x6 = np.linspace(1,10,25)
re_shaped = x6.reshape(5,5)   # Converting 1D array to 2D
re_shaped
```

# x) Flatten() method and ravel() method (Converting 2D to 1D)

```python
matrix = np.array([[1,7],[3,1]])
one_d = matrix.flatten()
one_d1 = matrix.ravel()
print(one_d)
print(one_d1)
```

# Reshape() examples

```python
test = np.array([1,2,3,5,9,7])
test.reshape(2,3)    # 2x3 matrix
test.reshape(3,2)    # 3x2 matrix
```
